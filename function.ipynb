6. Explain the concept of generators in Python and how they are defined.
  - In Python, a generator is a special type of function that produces a sequence of values one at a time, instead of returning a single value like a regular function. They are defined using the yield keyword instead of return.

     Generators are essentially iterators created using functions. They follow the iterator protocol by implementing the __iter__() and __next__() methods implicitly.

     The key advantage of generators is their lazy evaluation. They only generate values when needed, making them memory-efficient, especially when dealing with large datasets or infinite sequences.

     1. Using the yield keyword in a function:

            def my_generator(n):
            for i in range(n):
            yield i
            gen = my_generator(5)
            for value in gen:
            print(value)  # Output: 0, 1, 2, 3, 4

        In this example, my_generator is a generator function. When called, it returns a generator object gen. The yield keyword pauses the function's execution and returns a value. When the next value is requested, the function resumes from where it left off.

    2. Using generator expressions:

        Generator expressions are similar to list comprehensions but enclosed in parentheses () instead of square brackets []. They provide a concise way to create generators.


                gen = (i for i in range(5))
                for value in gen:
                print(value)  # Output: 0, 1, 2, 3, 4

7. What are the advantages of using generators over regular functions?
  - Memory Efficiency:

     Generators produce values on demand, meaning they only generate and store one value at a time. This is in contrast to regular functions, which typically compute and return the entire result set at once.

     This on-demand generation makes generators significantly more memory-efficient, especially when dealing with large datasets or infinite sequences. They avoid the need to store the entire sequence in memory, which can be crucial for resource-constrained environments.

 - Lazy Evaluation:

     Generators employ lazy evaluation, meaning they only compute and yield values when explicitly requested. This differs from regular functions, which execute all their code and return a result immediately.

     Lazy evaluation can lead to performance improvements, particularly when dealing with large datasets or complex computations. By generating values only when needed, generators avoid unnecessary calculations and resource consumption.

  - Representing Infinite Sequences:

     Generators can effortlessly represent infinite sequences, such as a stream of numbers or a continuous data feed. This is because they generate values one by one, without needing to store the entire sequence in memory.

     Regular functions, on the other hand, would struggle to represent infinite sequences as they would require infinite memory to store the entire sequence.

  - Improved Readability:

     In certain scenarios, generators can make code more concise and readable compared to regular functions. This is especially true when dealing with iterative processes or data pipelines.

     By using the yield keyword, generators encapsulate the logic of generating values, making the code more focused and easier to understand.

  - Pipeline Processing:

     Generators excel in pipeline processing, where data is transformed and passed through a series of operations.

     They enable the creation of efficient data pipelines by generating values on demand and passing them to the next stage of the pipeline without storing intermediate results in memory.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - These three functions are higher-order functions, meaning they take other functions as arguments and apply them to iterables. They are fundamental tools in functional programming and provide elegant ways to process and transform data.

     Here's a breakdown of their differences:

- map()

Purpose: Applies a given function to each item of an iterable and returns an iterator containing the results.

Transformation: Transforms each element of the iterable based on the provided function.
Output: An iterator with the same length as the input iterable, containing the transformed values.

Example:

             numbers = [1, 2, 3, 4, 5]
             squared_numbers = list(map(lambda x: x**2, numbers))  # Output: [1, 4, 9, 16, 25]

- reduce()

Purpose: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value.

Aggregation: Aggregates or combines the elements of the iterable into a single result.
Output: A single value representing the cumulative result of the function applied to the iterable.

Example:

                  from functools import reduce
                  numbers = [1, 2, 3, 4, 5]
                  product = reduce(lambda x, y: x * y, numbers)  # Output: 120

- filter()

Purpose: Constructs an iterator from elements of an iterable for which a function returns true.

Selection: Selects elements from the iterable based on a condition specified by the function.

Output: An iterator containing only the elements for which the function returned True.

Example:

                   numbers = [1, 2, 3, 4, 5]
                   even_numbers = list(filter(lambda x: x % 2 == 0, numbers))  # Output: [2, 4]

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


class FibonacciIterator:
    def __init__(self, limit):
        self.limit = limit  # The number of Fibonacci terms to generate
        self.current = 0   # The current term in the sequence
        self.first = 0    # First number in the Fibonacci sequence
        self.second = 1   # Second number in the Fibonacci sequence
        self.count = 0    # Number of terms generated so far

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.count < self.limit: # Check if we have not exceeded the limit
          if self.count == 0:      # Handle the first term case
            self.count += 1
            return 0
          elif self.count == 1:    # Handle the second term case
            self.count += 1
            return 1
          else:                  # Generate the next term
            next_fib = self.first + self.second
            self.first = self.second
            self.second = next_fib
            self.count += 1
            return next_fib
        else:
            raise StopIteration  # Raise StopIteration when the limit is reached

# Example usage:
fibonacci_sequence = FibonacciIterator(10)
for number in fibonacci_sequence:
    print(number)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file: # use a context manager to open and close file
        for line in file:
            yield line.strip() # removes trailing newline characters

# Example usage:
# Create a dummy file for testing
with open("dummy_file.txt", "w") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")

for line in read_file_line_by_line("dummy_file.txt"):
    print(line)

!rm dummy_file.txt

```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    vowels = "aeiouAEIOU"
    filtered_string = "".join(filter(lambda c: c not in vowels, string))
    return filtered_string


# Example Usage
my_string = "Hello World"
string_without_vowels = remove_vowels(my_string)
print(string_without_vowels)

```

```python
# 11) Imagine an accounting routine used in a book shop.
# It works on a list with sublists, which look like this:
#
# Order Number   Book Title and Author                  Quantity   Price per Item
# 34587          Learning Python, Mark Lutz            4          40.95
# 98762          Programming Python, Mark Lutz         5          56.80
# 77226          Head First Python, Paul Barry         3          32.95
# 88112          Einführung in Python3, Bernd Klein    3          24.99
#
# Write a Python program, which returns a list with 2-tuples.
# Each tuple consists of the order number and the product of the price per item and the quantity.
# The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.
#
# Write a Python program using lambda and map.

orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

def calculate_order_total(order):
  order_number, _, quantity, price = order
  total_price = quantity * price
  if total_price < 100:
      total_price += 10
  return (order_number, total_price)


order_totals = list(map(calculate_order_total, orders))
print(order_totals)
```
